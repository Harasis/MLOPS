```python
import seaborn as sns
```

```python
sns.set()
```

```python
tips = sns.load_dataset("tips")
#dataset given by seaborn only
```

```python
tips.columns
```

```python
tips.head(2)
```

```python
bill = tips['total_bill']
```

```python
bill.max()
```

```python
bill.min()
```

```python
sns.distplot(bill, kde = False)
```

```python
sns.distplot(bill, kde = False, bins = 50)
```

```python
sns.jointplot(data = tips , x = 'total_bill', y = 'tip')

#by default ScatterPlot
```

```python
sns.jointplot(data = tips , x = 'total_bill', y = 'tip', kind = 'reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex')

#sex wise distribution
```

```python
sns.barplot(data = tips, x = 'sex', y= 'total_bill')
```

```python
sns.countplot(data = tips, x= 'sex')
#to count the number of males and females
```

```python
sns.boxplot( data=tips, x='day', y='total_bill')
```

```python
sns.boxplot( data=tips, x='day', y='total_bill' , hue = 'sex')


#gender wise
```

```python
sns.violinplot( data=tips, x='day', y='total_bill' , hue = 'sex')
```

```python
sns.stripplot( data=tips, x='day', y='total_bill' , hue = 'sex')
```

```python
sns.stripplot( data=tips, x='day', y='total_bill' , hue = 'sex' , split = True, jitter=False)

#to give some looks and arrangements
```

```python
sns.swarmplot( data=tips, x='day', y='total_bill' , hue = 'sex')
```
