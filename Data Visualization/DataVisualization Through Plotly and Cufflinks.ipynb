```python
# Now Creating Plotly Interactive Graph
```

```python
import plotly.graph_objs as go
```

```python
import plotly.offline as pyo  #to create offline graphs
```

```python
datax = np.random.randint(1,101,100)
datay = np.random.randint(1,101,100)
```

```python
#graph_data = go.Scatter(x = datax, y= datay , mode = "marker")

graph_data =  [ go.Scatter(x=datax , y=datay, mode='markers' )  ]
```

```python
layout = go.Layout(title = "MY GRAPH" , hovermode= 'closest')
```

```python
fig = go.Figure(data= graph_data, layout= layout)
```

```python
pyo.plot(fig , filename='mygraph.html')
```
